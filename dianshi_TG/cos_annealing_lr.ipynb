```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cos_annealing_lr(initial_lr, cur_epoch, epoch_per_cycle):
    return initial_lr * (np.cos(np.pi * cur_epoch / epoch_per_cycle) + 1) / 2
```

```python
init_lr = 0.001
epoch_per_cycle = 10
lr_list = []
cycle_mult = 20
for cycle in range(3):
    epoch_per_cycle += cycle_mult
    for epoch in range(int(epoch_per_cycle)):
        lr_list.append(cos_annealing_lr(init_lr, epoch, epoch_per_cycle))
    cycle_mult *= 1.4
```

```python
plt.plot(lr_list)
plt.xlabel('epoch')
plt.ylabel('lr')
plt.show()
```
